# Veri Görselleştirme (Grafik)

```python
import matplotlib.pyplot as plt

plt.hist(df["Age"].dropna(), bins=20)
plt.title("Yas Dagilimi")
plt.show()
```

## Grafik 1 Yorumu

Yolcuların büyük kısmının 20-40 yaş aralığında olduğu görülmektedir.

```python
df["Sex"].value_counts().plot(kind="bar")
plt.title("Cinsiyet Dagilimi")
plt.show()
```

## Grafik 2 Yorumu
Erkek yolcu sayısı kadın yolcu sayısından fazladır.

```python
import seaborn as sns

sns.boxplot(x=df["Fare"])
plt.title("Bilet Ucretleri")
plt.show()
```

## Grafik 3 Yorumu
Bazı yolcuların bilet ücretleri diğerlerine göre oldukça yüksektir. Bilet fiyatlarının aynı oranda olduğunu söyleyemeyiz.

```python
plt.scatter(df["Age"], df["Fare"])
plt.xlabel("Yas")
plt.ylabel("Bilet Ucreti")
plt.title("Yas ve Bilet Ucreti")
plt.show()
```

## Grafik 4 Yorumu
Yaş ve bilet ücreti arasında güçlü bir ilişki görülmemektedir. Eşit oranda diyebiliriz.

```python
numeric_df = df.select_dtypes(include="number")

sns.heatmap(numeric_df.corr(), annot=True)
plt.title("Korelasyon Matrisi")
plt.show()
```

## Grafik 5 Yorumu
Burada sayısal değişkenler arasındaki ilişkiler korelasyon matrisi ile incelenmiştir.

## Bulgular 

- Erkek yolcu sayısı kadın yolcu sayısından fazladır.
- Yolcuların çoğu 20-40 yaş aralığındadır.
- Kadınların hayatta kalma oranı erkeklere göre daha yüksektir.
- Yaş ile bilet ücreti arasında belirgin bir ilişki gözlemlenmemiştir.
- Veri setinde yüksek ücretli bilet alan bazı yolcular bulunmaktadır.
